## Условие 5: 5 задача

###### 5.1 Выберите дом клиенту

###### Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?

```python
houses[(houses['waterfront']==True) & (houses['bathrooms']>=3) & (houses['sqft_basement']>0)]
```

###### 5.2 Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?

```python
df = houses[((houses['waterfront']==True) | (houses['view']==4)) & (houses['condition']==5) & (houses['yr_built']>=1980)]
df
```

```python
df.price.min()
```

```python
df.price.max()
```
