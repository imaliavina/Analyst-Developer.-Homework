### 1.1. Скачать данные по ссылке https://www.kaggle.com/datasets/ionaskel/laptop-prices

```python
import pandas as pd
import numpy as np

houses = pd.read_csv('./kc_house_data.csv')
houses.head()
```

### 1.2 Изучите стоимости недвижимости

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6, 4))

sns.histplot(houses['price'], bins=15)

plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость недвижимости')
plt.ylabel('Кол-во представленных домов')
plt.xticks(rotation=30);
```

Видим, что наибольшее кол-во домов находится по стоимости в диапазоне [0; 500000] - более 14000 позиций.  
Затем, в диапазоне [500000; 1000000] - около 5000 позиций.  
[1000000; 1500000] - чуть меньше 1000.  
Самые дорогие дома - [3000000; 8000000] - по количеству минимальны, их даже не видно на графике.  

Посмотрим, сколько домов из самой дорогой категории расположены по стоимости в этом диапазоне -  [3000000; 8000000].  

```python
houses[(houses['price']>=3000000) & (houses['price']<=8000000)]
```

Всего лишь 50 позиций. Судя по данным, они содержат наибольшее кол-во комнат (3-7), самую высокую оценку (11-13) и пр. параметры.

Найдем самую высокую цену дома:

```python
houses.price.max()
```

Посмотрим на цены более детально, отбросив 50 наиболее дорогих домов.

```python
df = houses[houses['price'] <= 3000000]

plt.figure(figsize=(6, 4))

sns.histplot(df['price'], bins=15)

plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость недвижимости')
plt.ylabel('Кол-во представленных домов')
plt.xticks(rotation=30);
```

Видим, что максимальное число домов находится по стоимости от 250000 до 500000.

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6, 4))

sns.histplot(df['sqft_living'], bins=15)

plt.title('Квадратура жилой площади')
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Кол-во представленных домов')
plt.xticks(rotation=30);
```

Видим, что наибольшее кол-во домов находится по стоимости в диапазоне [1000; 2000] - около 9000 позиций.
Затем, в диапазоне [2000; 3000] - около 6000 позиций.
[0; 1000] - чуть меньше 3000.
Самые обширные по площади дома - [6000; 14000] - по количеству минимальны, их даже не видно на графике.

Посмотрим, сколько домов расположены по площади в этом диапазоне - [6000; 14000].

```python
df1 = houses[(houses['sqft_living']>=6000) & (houses['sqft_living']<=14000)]
df1
```

68 домов. Посмотрим на средние характеристики для цены, оценки, кол-ва комнат для этих домов:

```python
print(f'Средняя цена: {df1.price.mean()}')
print(f'Среднее кол-во комнат: {df1.bedrooms.mean()}')
print(f'Средняя оценка: {df1.grade.mean()}')
print(f'Есть ли вид на море? : {df1.waterfront.mean()}')
print(f'Сколько этажей? : {df1.floors.mean()}')
```

Посмотрим на разброс по площади более детально, отбросив 68 наиболее обширных домов.

```python
df2 = houses[houses['sqft_living'] <= 6000]

plt.figure(figsize=(6, 4))

sns.histplot(df2['sqft_living'], bins=15)

plt.title('Квадратура жилой площади')
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Кол-во представленных домов')
plt.xticks(rotation=30);
```

### 2.1 Изучите распределение домов от наличия вида на набережную

- Сделайте выводы  

```python
df3 = houses['waterfront'].value_counts()
df3.index = ['no', 'yes']
df3.head()
```

```python
plt.figure(figsize=(6, 4))

plt.pie(df3, autopct='%1.1f%%')

plt.title('Доля домов с видом на набережную')
plt.legend(df3.index);
```

ПРоцент домов с видом на набережную составляет 0,8.

```python
df4 = pd.crosstab(df['grade'], df['waterfront']).reset_index().sort_values(by=0, ascending=False)
df4.rename(columns={0: 'no', 1: 'yes'}, inplace=True)
df4
```

```python
# Строим столбчатую многорядную диаграмму

plt.figure(figsize=(8, 6))

# определяем кол-во делений
n_ticks = np.arange(len(df4['no']))

# определяем сдвиг
offset = 0.2

# определяем ширину столбцов
w = 0.4

# добавляем сдвиг к кол-ву делений
plt.bar(n_ticks - offset, df4['no'], width=w)
plt.bar(n_ticks + offset, df4['yes'], width=w)

plt.title('Распределение домов с видом на набережную и без, в зависимости от оценки дома')
plt.xlabel('Grade')
plt.ylabel('Кол-во')
plt.legend(['no', 'yes'])
plt.xticks(n_ticks, df4['grade'], rotation = 30);  # добавляем метки делений
```

### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

```python
plt.figure(figsize=(6, 4))

sns.scatterplot(x=houses['sqft_living'], y=houses['price'])

plt.title('Зависимость стоимости недвижимости от жилой площади')
plt.xlabel('жилая площадь')
plt.ylabel('Стоимость ');
```

Видна довольно четкая линейная зависимость, хотя мы видим, что диапазон для выбора коэф-та угла наклона довольно широкий

```python
plt.figure(figsize=(6, 4))

sns.scatterplot(x=houses['bedrooms'], y=houses['price'])

plt.title('Зависимость стоимости недвижимости от кол-ва комнат')
plt.xlabel('Кол-во комнат')
plt.ylabel('Стоимость ');
```

Сложно интерпретировать график, видим, что дома с большим количеством комнат (например, 7) могут стоить столько же, сколько дома с 3 комнатами. Поэтому сложно сделать вывод.

```python
plt.figure(figsize=(6, 4))

sns.scatterplot(x=houses['bathrooms'], y=houses['price'])

plt.title('Зависимость стоимости недвижимости от кол-ва ванных комнат')
plt.xlabel('кол-во ванных комнат')
plt.ylabel('Стоимость ');
```

Здесь мы видим более выраженную зависимость, линейную.

Из графика видно, что дома в состоянии "3" могут стоить дороже, чем в состоянии "4" и "5". Это довольно неожиданно и необъяснимо. И требует дальнейшего анализа.

```python
plt.figure(figsize=(6, 4))

sns.scatterplot(x=houses['grade'], y=houses['price'])

plt.title('Зависимость стоимости недвижимости от оценки')
plt.xlabel('оценка')
plt.ylabel('Стоимость ');
```

С ростом оценки стоимость недвижимости растет, что вполне ожидаемо.

Построим тепловую карту основных признаков, чтобы лучше оценить их взаимосвязь. 

```python
df5 = houses[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'condition', 'grade', 'floors', 'waterfront', 'view']]
#df5.head()
sns.heatmap(df5.corr(), annot=True, linewidth=.5, cmap=sns.cubehelix_palette(as_cmap=True))
```

Видим, что цена недвижимости больше всего коррелирует с квадратурой, оценкой и кол-вом ванных комнат.
